# **Playgrounds in Pittsburgh**

This dataset contains all the playgrounds in Pittsburgh. The idea is that more playgrounds in a neighborhood makes it a better place to live. The data was last updated in 2021.

```python
import pandas as pd
import numpy as np

# load data from CSV using '_id' column as index
playgrounds = pd.read_csv("playgrounds.csv", index_col="_id")

# Count playgrounds by neighborhood and format as table
(playgrounds["neighborhood"]
    .value_counts()
    .rename_axis("Neighborhood")
    .reset_index(name="Playgrounds")
)
```
