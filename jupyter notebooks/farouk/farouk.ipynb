Neighborhoods with most playgrounds

```python
import pandas as pd
import numpy as np


playgrounds = pd.read_csv("playgrounds.csv", index_col="_id")
playgrounds["neighborhood"].value_counts().head()
```

Playgrounds in Squirrel Hill South

```python
from IPython.display import HTML

squirrel = playgrounds[playgrounds["neighborhood"] == "Squirrel Hill South"]
html = "<div style='display:flex; gap:30px'>"

for _, row in squirrel.tail(4).iterrows():
    html += f"""
        <div>
            <h4>{row["name"]}</h4>
            <img 
                src="{row['image']}" 
                width="300"  
                style="border-radius: 8px" 
            >
        </div>
    """

html += "</div>"
display(HTML(html))
```

Map of playgrounds

```python
import folium

m = folium.Map(location=[40.4406, -79.9959], zoom_start=12)

folium.GeoJson("map.geojson", name="Playgrounds").add_to(m)
folium.LayerControl().add_to(m)

m
```
