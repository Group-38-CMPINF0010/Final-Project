## The introduction
Here in this notebook, we begin creating one submetric.
At the moment, we are examining the parking space availability and rates in different lots in Pittsburgh.


```python
import pandas as pd

# Load the parking dataset
parking = pd.read_csv("parkingspace_rates.csv")  # it's actually parking lots data

# Quick look at columns
print("Columns:", parking.columns.tolist())
print(parking.head())
```

### Parking dataset
This file contains information about parking lots in Pittsburgh.  
It includes columns such as `zone`, `spaces`, `rate`, and `type`.  
I’ll use these to determine which neighborhoods have better parking access.


```python
# Clean up zone names
parking['zone'] = parking['zone'].astype(str).str.strip().str.title()

# Focus on useful columns
parking_data = parking[['zone','spaces','rate','type']].copy()

print(parking_data.head())
```

### Cleaning
I cleaned the `zone` names and kept only the important columns: `zone`, `spaces`, `rate`, and `type`.  
This makes it easier to analyze parking availability and cost.


```python
# Largest lots
largest_lots = parking_data.sort_values('spaces', ascending=False).head(10)

# Average rate by type
avg_rate = parking_data.groupby('type')['rate'].mean()

print("Top 10 largest lots:")
print(largest_lots)

print("\nAverage rate by type:")
print(avg_rate)
```

### Parking availability
I examined the largest lots (those with the most spaces) and calculated average rates by type (on-street vs. off-street).  
This shows which areas have more parking capacity and whether they tend to be cheaper or more expensive.

My part of the project focuses on parking zones and parking rates.
