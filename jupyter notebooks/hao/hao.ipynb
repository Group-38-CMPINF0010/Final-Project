# Metric: City of Pittsburgh facilities

The number of public facilities of City of Pittsburgh.

```python
import pandas as pd  # Data manipulation library
import matplotlib.pyplot as plt  # Plotting library

csv_path = "City Facilities.csv"
facilities_df = pd.read_csv(csv_path)  # Load the facilities dataset into a DataFrame

neighborhood_counts = (
    facilities_df["neighborhood"]  # Select the neighborhood column
    .value_counts()  # Count how many facilities exist for each neighborhood
    .rename_axis("neighborhood")  # Label the index as neighborhood for clarity
    .reset_index(name="facility_count")  # Convert counts to a table and name the column
)

neighborhood_counts  # Display the ranked table
```

The chart below ranks neighborhoods by the number of facilities they host.

```python
plt.figure(figsize=(14, 7))  # Make the plot wide enough for neighborhood labels
plt.bar(
    neighborhood_counts["neighborhood"],
    neighborhood_counts["facility_count"],
    color="steelblue",
)  # Create a bar chart showing the count per neighborhood
plt.xticks(rotation=90)  # Rotate x-axis labels so they are legible
plt.xlabel("Neighborhood")  # Label the x-axis
plt.ylabel("Facility count")  # Label the y-axis
plt.title("City of Pittsburgh Facilities by Neighborhood")  # Add a descriptive title
plt.tight_layout()  # Prevent label clipping around the edges
plt.show()  # Render the bar chart
```

### **Conclusion**

The data shows that the Top 5 best neighborhood that has the most facilities are: Highland Park(30), Squirrel Hill South(29), Perry North(20), Brookline(19), and Strip District(15).
