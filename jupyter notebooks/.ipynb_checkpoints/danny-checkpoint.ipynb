```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("house_value.csv")
col = df.filter(like="300,000 to $399,999").columns[0]


#removes commas and ensures conver to numerics
df[col] = pd.to_numeric(df[col].astype(str).str.replace(",", ""), errors="coerce")

#sorting the neighborhoods from most units to least (in descending order)
data = df[df[col] > 0].sort_values(col, ascending=False)[["Neighborhood", col]]

#spaces out the labels on the y-axis
plt.figure(figsize=(8, max(4, len(data) * 0.25)))

#prints bars, color green
plt.barh(data["Neighborhood"], data[col], color="C2")

#converts to descending order (start with squirrel hill south)
plt.gca().invert_yaxis()

#x-axis label
plt.xlabel("Units ($300k - $399k)")

#print graph
plt.show()
```

Graph for the neighborhoods 

**Conclusion**

all in all it appears the most affordable homes in pittsburgh can be found in  Squirrel hill south. So if you are ready to start a family  might want to go there for a affordable nice house.
